# Standard Normal Distributions
Many standard statistical models operate under the assumption that your dataset is normally distributed (parametric). Let's review what that means and go over some ways to check for normality using python. 

## Discrete and Continuous Random Variables
Let's say we blindly pick a random variable X from our dataset. The possible values for X can be described as either discrete if X can only take on a certain number of distinct values, such as [1,77,12] and continuous if X can take on an infinite number of values. 

If X is discrete and can only take on a limited set of values, such as 1,77 and 12 from above, then we can calculate the probility that X is either 1, 77 or 12, called it's probability mass function. Examples of these types of distributions are Bernoulli, Binomial and Poisson. 
- `disrete_X = [1,77,12]`
- `P(X) = 1/3`

For a continuous X, the probability that X can equal any particular value is known as it's probability distribution. Here, we are interested in the probability that X can take on any particular value in the range of real numbers. 
- `continuous_X = ∞`  
- `P(X) = 0` Because X can take on any, or all, real values.

## Standard Normal Distribution
The most famous, and most widely used, probability distribution function is the Gaussian, or standard normal, distribution. 

A Gaussian distribution has the following properties: 
- Mean = 0
